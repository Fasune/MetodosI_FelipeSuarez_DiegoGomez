$f(a) + \sum_{i=1}^{m} \frac{\partial f(a)}{\partial x_i} (x_i - a_i)$

o de manera vectorial:

$f(a) + \nabla f(a) \cdot (x - a)$

$f(x, y) = x^2 - y^2 + 2x$

$P(x, y) = 2 + 4(x - 1) - 2(y - 1)= 4x - 2y$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

x = sym.Symbol('x')
y = sym.Symbol('y')

f=lambda x,y: x**2 - y**2 + 2*x 
p=lambda x,y: 4*x - 2*y

x = np.linspace(-2, 4, 400)
y = np.linspace(-2, 4, 400)
X, Y = np.meshgrid(x, y)
Z = f(X,Y)
Z_tangente = p(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.plot_surface(X, Y, Z_tangente, cmap='plasma', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
