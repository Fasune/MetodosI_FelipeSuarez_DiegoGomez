```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
def funcion(x):
    return np.absolute(x-2)
```

```python
Dx = lambda f,x,h=1e-5: (f(x+h) - f(x-h))/(2*h)
_x = np.linspace(0,4,50)
F = funcion(_x)
```

```python
def Minimizer(f, N=1000, gamma=0.01):
    
    r = np.zeros(N)
   
    r[0] = np.random.uniform(-5,5)
    
    for i in (range(1,N)):
        r[i] = r[i-1] - gamma*Dx(f,r[i-1])
        
    return r
```

```python
x=Minimizer(funcion)
cost=funcion(x)
```

```python
plt.plot(_x,F)
plt.scatter(x[-1],cost[-1], color='r')
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.grid()
```

Se observa que a pesar de que f(x) no es diferenciable, con el metodo del gradiente se puede encontrar el minimo de la función. Lo anterior es debido a que se definió el gradiente a partir de la derivada numérica de la función, de ahí que no se presenten inconvenientes.
