```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = np.linspace(0,8,10)

y = np.zeros((3,len(x)))

y[0] = 2*x-2
y[1] = -0.5*x + 0.5
y[2] = 4 - x
```

```python
A = np.array([[2,-1],[1,2],[1,1]])
b = np.array([2,1,4])
At = A.T
print(At)
```

```python
M = np.dot(At,A)
bt = np.dot(At,b)
xsol = np.linalg.solve(M,bt)
#print(xsol)
xsol
```

```python
for l in range(y.shape[0]):
    plt.plot(x,y[l],ls='--',lw=2)
plt.scatter(xsol[0],xsol[1],color='r')
```

```python
distancia1=np.linalg.norm(np.dot(A,xsol)-b)
distancia1
```

```python
min_distance=float('inf')
for i in np.arange(-5,5,0.01):
    for j in np.arange(-5,5,0.01):
        _x=np.array([i, j])
        dist=np.linalg.norm(np.dot(A,_x)-b)
        if dist < min_distance:
            min_distance = dist
            min_point = (i, j)
print("Punto con la menor distancia:", min_point)
print("Menor distancia:", min_distance)
```

```python
plt.scatter(xsol[0],xsol[1],color='r', label='Solución MinCuadrados')
plt.scatter(min_point[0],min_point[1],color='b', label='Método iterativo')
plt.legend()
plt.xlim([1,2])
plt.ylim([0.4,0.5])
plt.grid()
```
