Calcular la probabilidad que n personas (n ≤ 365) tengan fechas diferentes de cumplea˜nos,
i.e, escribir la formula general de c´alculo. Grafique la probabilidad P(n ≤ 80) como
funci´on de n. Los n´umeros son demasiado grandes, pero Python puede manejar dichas
cantidades

```python
import matplotlib.pyplot as plt
import numpy as np
def prob_mismo_dia(n):
    list_cump=[]
    for i in range(n):
        list_cump.append((365 - i) / 365)

    prob_mismo_dia=np.prod(np.array(list_cump))
    return prob_mismo_dia

# Calcular la probabilidad para n de 1 a 80
n_vals = range(1, 81)
probabilidades=[]
for i in n_vals:
    probabilidades.append(prob_mismo_dia(i))

# Graficar la probabilidad
plt.plot(n_vals, probabilidades )
plt.xlabel('N')
plt.ylabel('Prob. dos personas')
plt.title('Probabilidad del Problema del Cumpleaños')
plt.grid(True)
plt.show()
```
