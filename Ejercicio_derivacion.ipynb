```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion1(x):
    return np.sqrt(np.tan(x))
```

```python
def derivadaReal(x):
    return (np.cos(x)**(-2))/(2*np.sqrt(np.tan(x)))
```

```python
def derivada(f):
    h=0.01
    return (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))
```

```python
x=np.linspace(0.1,1.1,100)
y=funcion1(x)
dy=derivada(funcion1)
plt.plot(x,y,label='función original')
plt.plot(x,dy,label='Derivada analítica')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
```

```python
def derCentral(f):
    h=0.01
    return (f(x+h)-f(x-h))/(2*h)
```

```python
dy=derivada(funcion1)
dyCentral=derCentral(funcion1)
dyReal=derivadaReal(x)
plt.plot(x,dy,label='Derivada prog orden O($h^{2}$)')
plt.plot(x,dyCentral,label='Derivada central orden O($h^{2}$)')
plt.plot(x,dyCentral,label='Derivada analítica')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
```

```python
errorprog=abs(dyReal-dy)
errorcentral=abs(dyReal-dyCentral)
plt.plot(x,errorprog,label='Error derivada progresiva')
plt.plot(x,errorcentral,label='Error derivada Central')
plt.legend()
plt.xlabel('x')
plt.ylabel('Error absoluto')
plt.grid()
```

Se puede concluir que se tiene el mismo orden de precisión ambos resultados (h^2).
Porque con un h=0.01 ambas aproximaciones presentan errores del orden de 10^-4. Aun así, la derivada central es más precisa
